```python
# Sum of digits in decadic number system
# 123 -> 1+2+3 -> 6

n = int(input())
s = 0

while n > 0:
    c = n % 10
    s += c
    n = n // 10

print(s)
```

```python
# Compute balistic curve of ball shot by connon. Cannon shots in direction of positive axis x with angle 80 from ground.
# Cannon is positioned at 0,0  

dt = 0.1
g = 9.81
x = 0.0
y = 0.0
t = 0.0
power = 50
angle = 80
vx = math.cos(math.radians(angle)) * power
vy = math.sin(math.radians(angle)) * power
xs = [] # !
ys = [] # !
while t < 10:
    print("position", x,y)
    xs.append(x) # !
    ys.append(y) # !
    x += vx * dt
    y += vy * dt
    vy -= g * dt
    t += dt

draw_fn(xs, ys)
```
